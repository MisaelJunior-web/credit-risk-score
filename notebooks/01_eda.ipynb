# EDA — Credit Risk Score

Análise exploratória do dataset *Give Me Some Credit*: distribuição da inadimplência,
relação com renda, idade, histórico de atraso e comprometimento de renda.

```python
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from data_prep import run as prepare_data

sns.set_style('whitegrid')
df = prepare_data()
df.head()
```

```python
# Taxa geral de inadimplência
df['SeriousDlqin2yrs'].value_counts(normalize=True)
```

```python
# Inadimplência por faixa etária
df['age_band'] = pd.cut(df['age'], bins=[18, 30, 40, 50, 60, 100])
df.groupby('age_band')['SeriousDlqin2yrs'].mean().plot(kind='bar', title='Taxa de inadimplência por faixa de idade')
plt.ylabel('Taxa de inadimplência')
plt.show()
```

```python
# Inadimplência por faixa de renda
df['income_band'] = pd.qcut(df['MonthlyIncome'], q=5, duplicates='drop')
df.groupby('income_band')['SeriousDlqin2yrs'].mean().plot(kind='bar', title='Taxa de inadimplência por faixa de renda')
plt.ylabel('Taxa de inadimplência')
plt.show()
```

```python
# Correlação entre features numéricas e o target
corr = df.select_dtypes(include=[np.number]).corr()['SeriousDlqin2yrs'].sort_values(ascending=False)
corr
```

```python
# Histórico de atraso total x inadimplência
sns.boxplot(data=df, x='SeriousDlqin2yrs', y='total_past_due')
plt.title('Total de atrasos x Inadimplência')
plt.show()
```
